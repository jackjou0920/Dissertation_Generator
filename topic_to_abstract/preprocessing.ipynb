```python
import os, re
import nltk
from nltk.tokenize import WordPunctTokenizer
```

```python
def save_file(path, text):
    with open(path, 'w', encoding='utf-8') as fp:
        fp.write(text)
```

```python
def normalise(text):
    lr = ""
    lines = text.splitlines()
    words = set(nltk.corpus.words.words())
    for line in lines:
        line = re.sub('\[(.)+\]|\((.)+\)', " ", line)
        line = re.sub('[^a-z0-9\,\s\.\[\]\(\)\?\:]', " ", line)
        
        title = re.match('\d+(\.)*(\d)*(\.)*(\d)*', line)
        figure = re.match('figure(\s)*\d+(\.)*(\d)*(\.)*(\d)*', line)
        table = re.match('table(\s)*\d+(\.)*(\d)*(\.)*(\d)*', line)
        stage = re.match('stage(\s)*\d+(\.)*(\d)*(\.)*(\d)*', line)
        chapter = re.match('chapter(\s)*(\d)*(\.)*(\d)*(\.)*(\d)*(\s)*(\:)*', line)
        chapter2 = re.match('chapter(\s)*(one|two|three)', line)
        if (title or figure or table or stage or chapter):
            continue
        # all numbers
        is_del = True
        for w in line.split():
            if (w.isdigit() == False):
                is_del = False    
        if (is_del):
            continue
        # only a alphabet
        num = 0
        for w in line.split():
            if (len(w) == 1 and w.isalpha()):
                num += 1
        if (num == 1):
            continue
        
        line = " ".join(w for w in nltk.wordpunct_tokenize(line) if w.lower() in words or not w.isalpha())
        lr += line + '\n'
    
#     print(lr)
    return lr
```

```python
def concate(text):
    word_punct_tokenizer = WordPunctTokenizer()
    tokens = word_punct_tokenizer.tokenize(text)
    text = " ".join(tokens)
    return text
```

```python
dirPath = "./msc_dataset/literature_review/"
desPath = "./msc_dataset/lr/"
files = [f for f in os.listdir(dirPath) if os.path.isfile(os.path.join(dirPath, f))]
files = sorted(files)
count = 0
for fname in files:
    if ("txt" not in fname):
        continue
    count += 1
    print(fname)
    text = ""
    with open(dirPath+fname, "r", encoding='utf-8') as fp:
        text = fp.read()
    text = normalise(text)
    text = concate(text)
#     print(text)
    save_file(desPath+fname, text)
#     if (count == 2):
#         break
```
